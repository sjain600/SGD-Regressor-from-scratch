```python
import numpy as np
```

```python
class SGDRegressor:
    """
    Constructor for SGDRegressor
    
    Parameters:
    learning_rate: the step size used in each update.
    epochs: No of passes over the training dataset
    batch_size: Number of samples to be used in each batch.
    reg (str): Type of regularization ('l1' or 'l2'); None if no regularization.
    reg_param (float): Regularization parameter.
    
    The weights and bias are initialized as None and will be set during the fit method.
        
    """
    
    def __init__(self, learning_rate=0.01, epochs=50, batch_size=1, reg=None, reg_param=0.0):
        
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        self.reg = reg
        self.reg_param = reg_param
        self.weights = None
        self.bias = None
        
    def fit(self, X, y):
        """Fit the SGDRegressor to the training data"""
        m, n = X.shape # m is number of samples, n is number of features
        self.weights = np.zeros(n)
        self.bias = 0
        
        # in each epoch, the data is shuffled, and batches are created to update the model parameters using SGD
        # outer loop
        for _ in range(self.epochs):  
            indices = np.random.permutation(m) # creates a random ordering of integers from 0 to m-1, stochastic part of SGD
            # X and y use the same indices 
            X_shuffled = X[indices]
            y_shuffled = y[indices]
            
            # inner loop
            for i in range(0, m, self.batch_size):
                X_batch = X_shuffled[i: i + self.batch_size]
                y_batch = y_shuffled[i: i + self.batch_size]
                
                pred = np.dot(X_batch, self.weights) + self.bias
                loss = y_batch - pred
                
                gradient_w = np.dot(X_batch.T, loss) / self.batch_size
                gradient_b = -2 * np.sum(loss) / self.batch_size
                
                if self.reg == 'l1':
                    gradient_w += self.reg_param * np.sign(self.weights)
                if self.reg == 'l2':
                    gradient_w += self.reg_param * self.weights
                
                self.weights -= self.learning_rate * gradient_w
                self.bias -= self.learning_rate * gradient_b
                
    
    def predict(self, X):
        """
        Returns predicted values using the linear model
        """
        return np.dot(X, self.weights) + self.bias
    
    def compute_loss(self, X, y):
        return (np.mean((y - self.predict(X)) ** 2) + self._get_regularization_loss()) ** 0.5
    
    def _get_regularization_loss(self):
        if self.reg == 'l1':
            return self.reg_param * np.sum(np.abs(self.weights))
        elif self.reg == 'l2':
            return self.reg_param * np.sum(self.weights ** 2)
        else:
            return 0
        
    def get_weights(self):
        """Returns the weights of the model"""
        return self.weights
        
        
        
```
